# Analiza sprzedaży gier wideo
**Cel projektu:** Podstawowa analiza i wizualizacja danych dotyczących sprzedaży gier wideo na świecie. 
**Źródło danych:** Zbiór `vgsales.csv` z platformy Kaggle.

```python
# Import potrzebnych bibliotek
!pip install pandas matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie danych
df = pd.read_csv('vgsales.csv')

# CZYSZCZENIE DANYCH: 
# Sprawdzamy rozmiar przed czyszczeniem
print("Rozmiar przed czyszczeniem:", df.shape)

# Usuwamy wszystkie wiersze, w których brakuje jakichkolwiek informacji (tzw. wartości NaN)
df_czyste = df.dropna()

# Sprawdzamy rozmiar po czyszczeniu
print("Rozmiar po czyszczeniu:", df_czyste.shape)
```

# Wizualizacja danych
Poniżej znajdują się dwa proste wykresy: jeden pokazujący najpopularniejsze gatunki gier, a drugi porównujący globalną sprzedaż w różnych latach.

```python
# WYKRES 1: Liczba wydanych gier według gatunku
plt.figure(figsize=(10, 5))
df_czyste['Genre'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Liczba wydanych gier z podziałem na gatunki')
plt.xlabel('Gatunek gry')
plt.ylabel('Ilość gier')
plt.show()

# WYKRES 2: Top 5 platform z największą liczbą gier
plt.figure(figsize=(8, 4))
df_czyste['Platform'].value_counts().head(5).plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Top 5 platform z największą liczbą gier')
plt.xlabel('Platforma (Konsola/PC)')
plt.ylabel('Ilość gier')
plt.show()
```

# Wnioski z analizy:
1. Po oczyszczeniu danych z pustych wartości, zbiór nadal pozostaje na tyle duży, że pozwala na wiarygodną analizę.
2. Jak widać na pierwszym wykresie, gatunkiem gier, którego wydano najwięcej, jest **Action** (Akcja).
3. Na drugim wykresie widać, że historycznie najwięcej gier w tym zestawieniu pojawiło się na konsolę **DS** oraz **PS2**.
